```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime
import cat_func as kt
import timeit

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
l_temp, l_city, l_Date, l_humidity, l_lat, l_lng, l_Cloudiness, l_Country, Max_temp, Wind_speed  = ([] for i in range(10))

kt.Kitty(1,0)
start = timeit.time.clock()
counter = 0
for i in range(len(cities)):
    counter = counter + 1
    if counter < 61: 
        print(f"Processing Record {i+1} | {cities[i]}")    
        # Build query URL and request your results in Celsius
        try:
            query_url = f"{url}appid={api_key}&units=imperial&q={cities[i]}"
            # Get weather data
            weather_response = requests.get(query_url).json()
            #Populate the lists of data
            l_temp.append(weather_response["main"]['temp'])
            l_city.append(weather_response["name"])
            l_Date.append(datetime.utcfromtimestamp(int(weather_response['dt'])).strftime('%Y-%m-%d %H:%M:%S'))
            l_humidity.append(weather_response["main"]['humidity'])
            l_lat.append(weather_response['coord']['lon'])
            l_lng.append(weather_response['coord']['lat'])
            l_Cloudiness.append(weather_response['clouds']['all'])
            l_Country.append(weather_response['sys']['country'])
            Max_temp.append(weather_response["main"]['temp_max'])
            Wind_speed.append(weather_response['wind']['speed'])
        except (KeyError, IndexError):
            kt.Kitty(2,1)
    elif 60-(timeit.time.clock()-start) > 0: 
            kt.Kitty(3,60-(timeit.time.clock()-start))
            counter = 0
            start = timeit.time.clock()
    else: 
            start = timeit.time.clock()
            counter = 0

kt.Kitty(4,1)
```

```python
#Create a DataFrame with all the information
df = {}
df.update({"City Name":l_city})
df.update({"Date":l_Date})
df.update({"Latitude":l_lat})
df.update({"Longitude":l_lng})
df.update({"Country":l_Country})
df.update({"Humidity":l_humidity})
df.update({"Cloudiness":l_Cloudiness})
df.update({"Max Temp":Max_temp})
df.update({"Wind Speed":Wind_speed})
df.update({"Temperature":l_temp})
pd.DataFrame(df).to_csv("CityData.csv",index=False)
```

```python
myPlots = ['Temperature', 'Humidity','Cloudiness','Wind Speed','Longitude']
fig, ax = plt.subplots()
for x in myPlots:
    y = 'Latitude'
    ax1 = pd.DataFrame(df)[[y,x]].plot(x=y, y=x, kind = 'scatter', title = f"{y} vs {x}")
    plt.savefig(f"{y}vs{x}.png")
```
